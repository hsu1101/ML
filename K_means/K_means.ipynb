```python
from sklearn.datasets import make_blobs
X,y = make_blobs(n_samples=150,n_features=2,centers=3,cluster_std = 0.5,random_state=0)
```

```python
X[:5]
```

```python
y
```

```python
import matplotlib.pyplot as plt
plt.scatter(X[:,0],X[:,1],c='white',marker='o',edgecolor ='black',s=50)

plt.grid()
plt.show()
```

### <font color =#000000>K_means </font>

```python
from sklearn.cluster import KMeans
#k值為n_clusters
km=KMeans(n_clusters=3,init='random',n_init=10,max_iter=300,tol=1e-04,random_state=0)

y_km = km.fit_predict(X)#y_km 為每個點屬於哪一個聚類
```

```python
y_km
```

```python
#如果y_km==0的話就拿X的第0個column
plt.scatter(X[y_km ==0,0],X[y_km==0,1],s=50,c='lightgreen',
           marker='s',edgecolor='black',label='cluster 1')
plt.scatter(X[y_km ==1,0],X[y_km==1,1],s=50,c='orange',
           marker='o',edgecolor='black',label='cluster 2')
plt.scatter(X[y_km ==2,0],X[y_km==2,1],s=50,c='lightblue',
           marker='v',edgecolor='black',label='cluster 3')

plt.scatter(km.cluster_centers_[:, 0],km.cluster_centers_[:,1],
           s=250,marker='*',c='red',edgecolor='black',label='centroids')

plt.legend(scatterpoints=1)
plt.grid()
plt.show()
```

### <font color =#000000>K_means++ </font>

```python
from sklearn.cluster import KMeans
#k值為n_clusters
km=KMeans(n_clusters=3,init='k-means++',n_init=10,max_iter=300,tol=1e-04,random_state=0)

y_km = km.fit_predict(X)#y_km 為每個點屬於哪一個聚類
```

```python
y_km
```

```python
plt.scatter(X[y_km ==0,0],X[y_km==0,1],s=50,c='lightgreen',
           marker='s',edgecolor='black',label='cluster 1')
plt.scatter(X[y_km ==1,0],X[y_km==1,1],s=50,c='orange',
           marker='o',edgecolor='black',label='cluster 2')
plt.scatter(X[y_km ==2,0],X[y_km==2,1],s=50,c='lightblue',
           marker='v',edgecolor='black',label='cluster 3')

plt.scatter(km.cluster_centers_[:, 0],km.cluster_centers_[:,1],
           s=250,marker='*',c='red',edgecolor='black',label='centroids')

plt.legend(scatterpoints=1)
plt.grid()
plt.show()
```

###  <font color =#000000>凝聚式階層聚類_complete linkage </font>

```python
from scipy.cluster.hierarchy import linkage
from scipy.cluster.hierarchy import dendrogram

Complete = linkage(X,method='complete',metric='euclidean')
dendrogram(Complete)
ax = plt.gca()
plt.xlabel("Sample index")
plt.ylabel("Cluster distance")
plt.show()
```

###  <font color =#000000>凝聚式階層聚類_single linkage </font>

```python
from scipy.cluster.hierarchy import linkage
from scipy.cluster.hierarchy import dendrogram

Complete = linkage(X,method='single',metric='euclidean')
dendrogram(Complete)
ax = plt.gca()
plt.xlabel("Sample index")
plt.ylabel("Cluster distance")
plt.show()
```

###  <font color =#000000>凝聚式階層聚類_average linkage </font>

```python
from scipy.cluster.hierarchy import linkage
from scipy.cluster.hierarchy import dendrogram

Complete = linkage(X,method='average',metric='euclidean')
dendrogram(Complete)
ax = plt.gca()
plt.xlabel("Sample index")
plt.ylabel("Cluster distance")
plt.show()
```

###  <font color =#000000>Wards Method  </font>

```python
from scipy.cluster.hierarchy import ward
from scipy.cluster.hierarchy import dendrogram

linkage = ward(X)
dendrogram(linkage)
plt.xlabel("Sample index")
plt.ylabel("Cluster distance")
plt.show()
```

```python
from sklearn.datasets import make_moons

X,y=make_moons(n_samples=200,noise=0.05,random_state=0)
plt.scatter(X[:,0],X[:,1])
plt.show()
```

```python
from sklearn.cluster import DBSCAN

db = DBSCAN(eps=0.2,min_samples=5,metric = 'euclidean')
y_db = db.fit_predict(X)
plt.scatter(X[y_db==0,0],X[y_db==0,1],c='lightblue',marker='o',s=40
            ,edgecolor='black',label='cluster 1')
plt.scatter(X[y_db==1,0],X[y_db==1,1],c='red',marker='s',s=40
            ,edgecolor='black',label='cluster 2')
plt.legend()
plt.show()
```
